```python
from tkinter import *
myw=Tk()
myw.geometry('480x480')
myw.configure(bg='cyan')
myw.title('Currency Converter')

#LABELS
label1=Label(myw,text="Convert From",width=10,height=2,bg='cyan')
label1.place(x=50,y=0)
label2=Label(myw,text="Convert To",width=8,height=2,bg='cyan')
label2.place(x=360,y=0)
label3=Label(myw,text='Converted Value',width=22,height=2,bg='white',borderwidth=1,relief="solid")
label3.place(x=160,y=300)
label4=Label(myw,text='Enter Value',width=22,height=2,bg='cyan')
label4.place(x=160,y=170)

#Entry
entry1=Entry(myw)
entry1.place(x=180,y=150)




#Dropdown menu options
options = [ "USD","INR","EUR","CAD","JPY"]

        
#datatype of menu text
drop1_variable=StringVar()
drop1_variable.set("Select")
drop2_variable=StringVar()
drop2_variable.set("Select")


# Create Dropdown menu
drop1=OptionMenu(myw,drop1_variable,*options)
drop1.place(x=50,y=50)
drop2=OptionMenu(myw,drop2_variable,*options)
drop2.place(x=360,y=50)


#The Convert Function
def myfunc():   
    userchoice1=float(entry1.get())
    print(userchoice1)
    print(str(drop1_variable.get()))
    print(str(drop2_variable.get()))
    
    if str(drop1_variable.get())=="USD":
        if str(drop2_variable.get())=="USD":
            label3.config(text=userchoice1*1)
        if str(drop2_variable.get())=="INR":
            label3.config(text=userchoice1*74.80)
        if str(drop2_variable.get())=="EUR":
            label3.config(text=userchoice1*0.86)
        if str(drop2_variable.get())=="CAD":
            label3.config(text=userchoice1*1.33)
        if str(drop2_variable.get())=="JPY":
            label3.config(text=userchoice1*107)
        
    
    elif str(drop1_variable.get())=="INR":
        if str(drop2_variable.get())=="USD":
            label3.config(text=userchoice1/74.80)
        if str(drop2_variable.get())=="INR":
            label3.config(text=userchoice1*1)
        if str(drop2_variable.get())=="EUR":
            label3.config(text=(userchoice1/74.80)*0.84)
        if str(drop2_variable.get())=="CAD":
            label3.config(text=(userchoice1/74.80)*1.33)
        if str(drop2_variable.get())=="JPY":
            label3.config(text=(userchoice1/74.80)*107)
            
    elif str(drop1_variable.get())=="EUR":
        if str(drop2_variable.get())=="USD":
            label3.config(text=userchoice1/0.84)
        if str(drop2_variable.get())=="INR":
            label3.config(text=(userchoice1/0.84)*74.80)
        if str(drop2_variable.get())=="EUR":
            label3.config(text=userchoice1*1)
        if str(drop2_variable.get())=="CAD":
            label3.config(text=(userchoice1/0.84)*1.33)
        if str(drop2_variable.get())=="JPY":
            label3.config(text=(userchoice1/0.84)*107)
    
    elif str(drop1_variable.get())=="CAD":
        if str(drop2_variable.get())=="USD":
            label3.config(text=userchoice1/1.33)
        if str(drop2_variable.get())=="INR":
            label3.config(text=(userchoice1/74.80)*1.33)
        if str(drop2_variable.get())=="EUR":
            label3.config(text=(userchoice1/0.84)*1.33)
        if str(drop2_variable.get())=="CAD":
            label3.config(text=userchoice1*1)
        if str(drop2_variable.get())=="JPY":
            label3.config(text=(userchoice1/1.33)*107)
            
    elif str(drop1_variable.get())=="JPY":  
        if str(drop2_variable.get())=="USD":
            label3.config(text=userchoice1/107)
        if str(drop2_variable.get())=="INR":
            label3.config(text=(userchoice1/107)*74.8)
        if str(drop2_variable.get())=="EUR":
            label3.config(text=(userchoice1/107)*0.84)
        if str(drop2_variable.get())=="CAD":
            label3.config(text=(userchoice1/107)*1.33)
        if str(drop2_variable.get())=="JPY":
            label3.config(text=userchoice1*1)
        
        

                   
#Convert Button
button1=Button(myw,text='Convert',command=myfunc,width=10,height=2,bg='yellow')
button1.place(x=200,y=240)

#Quit Button
button2=Button(myw,text='Quit',command=myw.destroy,fg='red',width=6,height=2)
button2.place(x=215,y=400)


mainloop()
```
